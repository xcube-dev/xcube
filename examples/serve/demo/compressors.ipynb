```python
import xarray as xr
import numpy as np
from zarr.storage import MemoryStore
from fsspec.implementations.asyn_wrapper import AsyncFileSystemWrapper
import fsspec
```

```python
# create dataset
data = xr.Dataset(
    {
        "B01": (("time", "lat", "lon"), np.random.rand(2, 3, 3)),
    },
    coords={
        "time": np.arange(2),
        "lat": np.arange(3),
        "lon": np.arange(3),
    },
)
data
```

```python
fs = fsspec.filesystem("memory")
fs = AsyncFileSystemWrapper(fs)
store = fs.get_mapper("dataset_1.zarr")
data.to_zarr(store, mode="w")
```

```python
# read back
ds = xr.open_zarr(store, consolidated=False)
ds
```
